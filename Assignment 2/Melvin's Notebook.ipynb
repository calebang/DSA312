Use the fitted KNN regression from `Q3a` to get the predicted value for the dummy observation, and calculate the corresponding loss function (i.e. Mean Squared Error) for this dummy observation.
##### (c) Use a validation approach where 70% of the data is used for training and 30% is used for validation. Perform a KNN regression using `K=3` on the training data. Calculate and report the MSE on the validation set.
##### (d) Perform hyperparameter tuning to find the optimal value of K for KNN, using K values ranging from 1 to 20, with 5-folds cross validation strategy. Use the fitted optimal KNN regression from `Q3e` to get the predicted value for the dummy observation, and calculate the corresponding loss function (i.e. Mean Squared Error) for this dummy observation.

```python
# Q3 Load Libraries Here
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Q3a
cali = fetch_california_housing()
X = cali.data
y = cali.target

scaler = StandardScaler()
X = scaler.fit_transform(X)

k = 3
knn_regressor = KNeighborsRegressor(n_neighbors = k)
knn_regressor.fit(X, y)

y_pred = knn_regressor.predict(X)
mse = mean_squared_error(y, y_pred)

print(f"MSE: {mse}")

# Answer:
# The corresponding MSE is 0.21576471308116868.
```

```python
# Q3b
observation = np.array([[5.0, 25, 6.0, 1.1, 1500, 3.0, 34.05, -118.25]])
observation = scaler.transform(observation)
observation_pred = knn_regressor.predict(observation)
mse2 = mean_squared_error([[2]], observation_pred)

print(f"Predicted Value: {observation_pred}")
print(f"Squared Error: {mse2}")

# Answer:
#The predicted value is 3.45067.
#The squared error on the observation is 2.1044434489000006.
```

```python
# Q3c
X = cali.data
y = cali.target
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 3, random_state = 8)

X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

knn_regressor.fit(X_train, y_train)
y_pred3 = knn_regressor.predict(X_val)
mse3 = mean_squared_error(y_val, y_pred3)

print(f"Validation MSE: {mse3}")

# Answer:
# The MSE on the validation set is 0.17123607407407415.
```

```python
# Q3d
pipeline = Pipeline([("scaler", StandardScaler()), ("knn", KNeighborsRegressor())])
param_grid = {
    "knn__n_neighbors": list(range(1,21))
}
kf = KFold(n_splits = 5, shuffle = True, random_state = 8)
grid_search = GridSearchCV(pipeline, param_grid, cv = kf, scoring = "neg_mean_squared_error")
grid_search.fit(X_train, y_train)

best_k = grid_search.best_params_["knn__n_neighbors"]
best_knn_regressor = grid_search.best_estimator_
y_pred4 = best_knn_regressor.predict(X_val)
mse4 = mean_squared_error(y_val, y_pred4)

print(f"Optimal number of neighbors (k): {best_k}")
print(f"Best cross-validated MSE for k={best_k}: {-grid_search.best_score_:.4f}\n")
print(f"Mean Squared Error (MSE) on validation set with k={best_k}: {mse4:.2f}")

# Answer:
# Optimal k is 11 and the corresponding CV MSE is 0.4038
# The MSE on the validation set is 0.05.
```

```python
# Q3e
final_knn_regressor = KNeighborsRegressor(n_neighbors = 11)
X = scaler.transform(X)
final_knn_regressor.fit(X, y)
y_pred5 = final_knn_regressor.predict(X)
mse5 = mean_squared_error(y, y_pred5)

print(f"Corresponding MSE: {mse5}")

# Answer:
# The corresponding MSE is 0.3265808044233437.
```
