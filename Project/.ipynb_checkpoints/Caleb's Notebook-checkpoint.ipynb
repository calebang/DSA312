```python
from scipy.stats import chi2_contingency

def chi_square_test(var1, var2):
    contingency_table = pd.crosstab(data[var1], data[var2])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    return chi2, p

# Example: Testing association between SMOKING and YELLOW_FINGERS
chi2, p_value = chi_square_test('SMOKING', 'YELLOW_FINGERS')
print(f"Chi-Square: {chi2}, p-value: {p_value}")
```

```python
data1
```

```python
variables = ['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
       'SWALLOWING DIFFICULTY', 'CHEST PAIN', 'LUNG_CANCER']
```

```python

# Initialize a list to store results
chi_square_results = []

for var in variables:
    # Create a contingency table
    contingency_table = pd.crosstab(data[var], data['LUNG_CANCER'])
    
    # Perform Chi-Square test
    chi2, p, _, _ = chi2_contingency(contingency_table)
    
    # Store the results
    chi_square_results.append((var, chi2, p))

# Convert results to a DataFrame for better readability
results_df = pd.DataFrame(chi_square_results, columns=['Variable', 'Chi-Square', 'p-value'])

# Display the results
results_df[results_df['p-value'] < 0.05]
```

```python
results_df[results_df['p-value'] > 0.05]
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
from itertools import combinations

# Initialize a list to store results
chi_square_results2 = []

# Loop through all combinations of variables
for var1, var2 in combinations(variables, 2):
    # Create a contingency table
    contingency_table = pd.crosstab(data[var1], data[var2])
    
    # Perform Chi-Square test
    chi2, p, _, _ = chi2_contingency(contingency_table)
    
    # Store the results
    chi_square_results2.append((var1, var2, chi2, p))

# Convert results to a DataFrame for better readability
results_df2 = pd.DataFrame(chi_square_results2, columns=['Variable 1', 'Variable 2', 'Chi-Square', 'p-value'])

# Display the results
results_df2
```

```python
results_df2[results_df2['p-value'] <= 0.05]
```

### Yellow Finger, smoking analysis

```python
import pandas as pd
```

```python
data = pd.read_csv("C://Users//drago//OneDrive//Desktop//UNI//Y4S1//DataSciPython//lung cancer survey.csv")
```

```python
data = data.dropna()
```

```python
data
```

```python
data.describe()
```

```python
import pandas as pd

contingency_smoking = pd.crosstab(data['SMOKING'], data['LUNG_CANCER'])
print(contingency_smoking)
```

```python
non_smokers_no_cancer = contingency_smoking.loc[0, 0]
non_smokers_with_cancer = contingency_smoking.loc[0, 1]

smokers_no_cancer = contingency_smoking.loc[1, 0] 
smokers_with_cancer = contingency_smoking.loc[1, 1]
```

```python
total_non_smokers = non_smokers_no_cancer + non_smokers_with_cancer
total_smokers = smokers_no_cancer + smokers_with_cancer

prop_non_smokers_with_cancer = non_smokers_with_cancer / total_non_smokers
print(f'Proportion of non-smokers with lung cancer: {prop_non_smokers_with_cancer:.2f}')

prop_smokers_with_cancer = smokers_with_cancer / total_smokers
print(f'Proportion of smokers with lung cancer: {prop_smokers_with_cancer:.2f}')
```

```python
contingency_yellow_fingers = pd.crosstab(data['YELLOW_FINGERS'], data['LUNG_CANCER'])
print(contingency_yellow_fingers)
```

```python
no_yellow_fingers_no_cancer = contingency_yellow_fingers.loc[0, 0]
no_yellow_fingers_with_cancer = contingency_yellow_fingers.loc[0, 1]

yellow_fingers_no_cancer = contingency_yellow_fingers.loc[1, 0] 
yellow_fingers_with_cancer = contingency_yellow_fingers.loc[1, 1]

total_no_yellow_fingers = no_yellow_fingers_no_cancer + no_yellow_fingers_with_cancer
total_yellow_fingers = yellow_fingers_no_cancer + yellow_fingers_with_cancer

prop_no_yellow_fingers_with_cancer = no_yellow_fingers_with_cancer / total_no_yellow_fingers
print(f'Proportion of people without yellow fingers with lung cancer: {prop_no_yellow_fingers_with_cancer:.2f}')

prop_yellow_fingers_with_cancer = yellow_fingers_with_cancer / total_yellow_fingers
print(f'Proportion of people with yellow fingers with lung cancer: {prop_yellow_fingers_with_cancer:.2f}')
```

### A large proportion of smokers have lung cancer. 83% of smokers have lung cancer. it seems that respondents with yellow fingers have 8% higher chance of getting lung cancer as opposed to those who just smoke. Those with yellow fingers imply that they are heavy smokers, when asked if people smoke, social smoking counts as =1 as well. In this aspect,  smoking is a clear factor of lung cancer and heavy smoking only increases your chances of getting lung cancer.

### Comparison of symptoms

```python
columns = ['FATIGUE', 'WHEEZING', 'SHORTNESS OF BREATH', 'CHEST PAIN']
```

```python
contingency_table = pd.crosstab([data['FATIGUE '], data['WHEEZING'],data['SHORTNESS OF BREATH'], data['CHEST PAIN'], data['COUGHING']], data['LUNG_CANCER'])
```

```python
print(contingency_table)
```

```python
contingency_table
```

```python
contingency_table['Total'] = contingency_table[0.0] + contingency_table[1.0]

contingency_table['Proportion_Lung_Cancer'] = contingency_table[1.0] / contingency_table[1.0].sum()

contingency_table['Proportion_No_Lung_Cancer'] = contingency_table[0.0] / contingency_table[0.0].sum()

print(contingency_table[['Proportion_Lung_Cancer', 'Proportion_No_Lung_Cancer']])
```

### Fatigue Alone May Not Be a Strong Indicator. While fatigue is a common symptom, it seems to contribute less to the increase in lung cancer proportions compared to other symptoms. The increase in lung cancer proportions is more significant when fatigue is combined with symptoms like wheezing, shortness of breath, or coughing.

### The more symptoms a person has (especially fatigue, wheezing, shortness of breath, chest pain, and coughing), the higher the proportion of lung cancer cases. For instance: In the row where all symptoms except wheezing are present (FATIGUE=1, SHORTNESS OF BREATH=1, CHEST PAIN=1, COUGHING=1), the proportion of lung cancer is 0.092465, which is among the highest. Compare this to a row with no symptoms (FATIGUE=0, WHEEZING=0, SHORTNESS OF BREATH=0, CHEST PAIN=0, COUGHING=0), where the proportion of lung cancer is much lower at 0.010075.

### Coughing appears to be a critical symptom associated with lung cancer. The rows with coughing (especially in conjunction with other symptoms) show consistently higher lung cancer proportions compared to rows without coughing.

```python
symptoms = ['FATIGUE ', 'WHEEZING', 'SHORTNESS OF BREATH', 'CHEST PAIN', 'COUGHING']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

proportions = data[symptoms + ['LUNG_CANCER']].groupby('LUNG_CANCER').mean().T

# Plot the stacked bar chart
proportions.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Proportional Stacked Bar Chart of Symptoms by Lung Cancer Status')
plt.xlabel('Symptoms')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Lung Cancer Status', labels=['No', 'Yes'])
plt.show()
```
